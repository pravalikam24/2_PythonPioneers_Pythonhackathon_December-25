```python
binary_cols = [
    'probable', 'vulnerable', 'is_most_recent',
    'fever_chills_shakes', 'cough', 'shortness_of_breath',
    'any_medical_conditions', 'travel_outside_canada',
    'contact_with_illness', 'symptoms', 'conditions', 'needs'
]
```

```python
yn_map = {'y': 1, 'n': 0}

for col in binary_cols:
    df2[col] = (
        df2[col]
        .astype(str)
        .str.lower()
        .str.strip()
        .map(yn_map)
    )
```

```python
df2[binary_cols].dtypes
```

```python
df2['cough'].value_counts(dropna=False)
```

```python
schema2_fsa = df2.groupby('fsa')[binary_cols].mean().reset_index()
schema2_fsa.head()
# Aggregate by FSA - creates one row per fsa
```
